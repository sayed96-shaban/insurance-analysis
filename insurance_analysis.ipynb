```python
# insurance_analysis.py
# Project: Medical Insurance Charges Analysis
# Libraries used: Python, Pandas, Numpy, Seaborn, Matplotlib

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# 1) Load Dataset

df = pd.read_csv("insurance.csv")   # make sure insurance.csv is in same folder
df
print("\n--- Dataset Sample ---")
print(df.head())

# 2) Basic Info

print("\n--- Info ---")
print(df.info())

print("\n--- Stats ---")
print(df.describe(include="all"))

# -------------------------
# 3) Distribution of Charges
# -------------------------
plt.figure(figsize=(8,5))
sns.histplot(df['charges'], kde=True, color="blue")
plt.title("Distribution of Insurance Charges")
plt.show()

# -------------------------
# 4) Charges by Smoker
# -------------------------
plt.figure(figsize=(6,5))
sns.boxplot(x="smoker", y="charges", hue="smoker", data=df, palette="Set2", legend=False)
plt.title("Charges by Smoking Status")
plt.show()

# -------------------------
# 5) Charges by Gender
# -------------------------
plt.figure(figsize=(6,5))
sns.boxplot(x="sex", y="charges",hue="smoker", data=df, palette="Set3",legend=False)
plt.title("Charges by Gender")
plt.show()

# -------------------------
# 6) Charges by Region
# -------------------------
plt.figure(figsize=(8,5))
sns.boxplot(x="region", y="charges",hue="smoker", data=df, palette="coolwarm",legend=False )
plt.title("Charges by Region")
plt.show()


# 7) Age vs Charges (Smoker highlighted)

plt.figure(figsize=(8,6))
sns.scatterplot(x="age", y="charges", hue="smoker", data=df, palette="Set1")
plt.title("Age vs Charges (Smoker vs Non-Smoker)")
plt.show()

# 8) BMI vs Charges

plt.figure(figsize=(8,6))
sns.scatterplot(x="bmi", y="charges", hue="smoker", data=df, palette="Dark2")
plt.title("BMI vs Charges")
plt.show()


# 9) Correlation Heatmap (Numeric only)

plt.figure(figsize=(6,5))
corr = df.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap="Blues", fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()


# 10) Insights

print("\n--- Insights ---")
print("1. Smokers pay significantly higher charges than non-smokers.")
print("2. Charges increase with age (older → higher charges).")
print("3. Higher BMI (overweight/obese) → higher charges, especially if smoker.")
print("4. Number of children has only a small effect on charges.")
print("5. Region and gender show minor differences compared to smoker & age.")
```
